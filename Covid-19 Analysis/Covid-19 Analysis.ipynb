### Data Modelling

We will fit our model according to given equation:
####   P_t = P_o * (e^(r*t)), where t is the time period in days, r is the rate of increase,
####                          P_t is the predicted forecast and P_o is the initial value.      
#### rate of increase, r = (No. of cases on next day - No. of cases on present day)/ No. of cases on present day

```python
# To compute rate of increase on each day from 04/03/2020 and 21/03/2020
data['Increased_Cases'] = 0
startdate = "2020-03-04"
enddate = "2020-03-21"
filtered_data = data[(data['Date'] >= startdate) & (data['Date'] <= enddate)]
```

```python
filtered_data.reset_index(drop=True,inplace=True)
```

```python
filtered_data
```

```python
r = 0
cases = 0
for i in range(0,filtered_data.shape[0]-1):
    daily_rate = (filtered_data['total cases'][i+1] - filtered_data['total cases'][i])/filtered_data['total cases'][i]
    r += daily_rate
r /= filtered_data.shape[0]-1
print(r)
```

**Finding the value of P_t when P_o=31 and t=26 and rounding of the value to nearest integral value**

```python
P_o, t = 31, 26
P_t = P_o * np.exp(r*t)
P_t = round(P_t)
P_t
```
