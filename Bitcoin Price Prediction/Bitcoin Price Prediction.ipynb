```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Bitcoin.csv')
data.shape
```

```python
data['btc_market_price'][1023]
```

```python
import seaborn as sns
plt.figure(figsize=(30,26))
sns.heatmap(data.corr(method='pearson'),annot=True)
plt.show()
```

```python
X = data[['btc_market_cap','btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent']]
Y = data['btc_market_price']
```

```python
print(X.shape,Y.shape)
```

```python
print(X.isna().sum())
```

```python
X.fillna(X.mean(),inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X.values,Y.values,test_size=0.25,random_state = 42)
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(X_train,Y_train)
print("Score on test set is: ",model.score(X_test,Y_test))
```

```python
y_pred = model.predict(X_test)
```

```python
print("Mean Squarred Error is: ",mean_squared_error(Y_test,y_pred))
```
